```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
# can also use from sklearn.linear_model import LinearRegression but I like statsmodels because of the way it gives nice output!
```

```python
# Load the data

path = "data/Advertising.csv"
data = pd.read_csv(path)

#df = pd.read_csv(path)[["Metropolitan", "White", "Graduates", "Poverty", "PercentFemale"]].dropna()

```

```python

# Define the model
model = smf.ols('sales ~ TV + radio + newspaper', data=data).fit()

# Get the summary as a DataFrame
results_df = pd.DataFrame({
    'Coefficient': model.params.round(3),
    'Standard Error': model.bse.round(3),
    't Value': model.tvalues.round(2),
    'P Value': model.pvalues.round(3)
})

# Add R-squared and Adjusted R-squared at the end of the table
r_squared = pd.DataFrame({'Coefficient': ['R-squared', model.rsquared, '', ''],
                          'Standard Error': ['Adjusted R-squared', model.rsquared_adj, '', ''],
                          't Value': ['', '', '', ''],
                          'P Value': ['', '', '', '']})

results_df = pd.concat([results_df, r_squared], ignore_index=True)


# Display the formatted table
results_df
```
